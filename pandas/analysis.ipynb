```python
import pandas as pd
data = {
    "Name": ['Ram', 'Ghanshyam', None, 'Aditi',None, 'Jagdish', 'Raj', 'Simran'],
    "Age": [28, 32, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, None, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performance_Score": [85, None, 78, 92, 88, 95, 80, 89]
}
df = pd.DataFrame(data)
df.dropna(inplace=True)
```

```python
df.sort_values(by='Name',ascending=False)
```

```python
df.sort_values(by=['Name','Salary'],ascending=[True,False])
```

```python
df.groupby('Age')['Salary'].sum()
```

```python
df.groupby(['Age','Name'])['Salary'].sum()
```

```python
import pandas as pd

# Create example DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value1': [1, 2, 3, 5]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D', 'E'], 'value2': [4, 5, 6, 7]})

# Perform an inner merge (default) on the 'key' column
merged_df_inner = pd.merge(df1, df2, on='key', how='inner')
print("Inner Merge:")
print(merged_df_inner)

# Perform an outer merge
merged_df_outer = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Merge:")
print(merged_df_outer)
```
